# Box created using shape.box

```python
%reset -f
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from mpp import plot_wires_3d, to_mpp
import shape


box = shape.box(
    x=1, y=0, z=2, l=1, w=2, h=3, phi=30, kwargs={'c': 'k'})

p_e = np.array([4, 1, 3])
v_e = np.array([-1, -.3, -.3])

ax = plot_wires_3d(box, p_e, v_e)[1]

ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_zlim(-5, 5)

phi = 0
a = 75
b = 75

to_mpp(box, p_e=p_e, v_e=v_e, phi=phi, a=a, b=b, width=10);
```

# Floating elliptic cylinder

```python
%reset -f
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from mpp import plot_wires_3d, to_mpp
import shape


cyl = shape.cyl(0, 0, 3, 1, 2, 1, 35, kwargs={'c': 'k'})

p_e = np.array([2.5, 0, 2])
v_e = np.array([-1, 0, .3])

ax = plot_wires_3d(cyl, p_e, v_e)[1]

ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
ax.set_zlim(-3, 3)

phi = 30
a = 100
b = 100

to_mpp(cyl, p_e=p_e, v_e=v_e, phi=phi, a=a, b=b, width=10);
```
