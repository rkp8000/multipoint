# Ellipsoid interior

```python
%reset -f
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from mpp import plot_wires_3d, to_mpp
import shape
```

```python
A = 6
C = 3
Z_0 = -1.5

wires = {}
```

```python
# z cross sections
for ctr, psi in enumerate(np.linspace(-np.pi/2, np.pi/2, 28)):
    z = C * np.sin(psi)
    w = []
    
    for theta in np.linspace(0, 2*np.pi, 500):
        
        num = 1 - (z**2)/(C**2)
        denom = (np.cos(theta)**2) / (A**2) + (np.sin(theta)**2) / (A**2)
        
        r = np.sqrt(num/denom)
        
        x = r*np.cos(theta)
        y = r*np.sin(theta)
        
        w.append([x, y, z])
    
    key = 'z_{}'.format(ctr)
     
    wires[key] = [np.array(w), {'c': 'g'}]
```

```python
# bottom cross section
w = []

for theta in np.linspace(0, 2*np.pi, 500):
    r = np.sqrt(A**2 * (1 - Z_0**2/C**2))
    
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    z = Z_0
    
    w.append([x, y, z])

wires['bottom'] = [np.array(w), {'c': 'g'}]
```

```python
# rotational cross sections
for ctr, alph in enumerate(np.linspace(0, 2*np.pi, 29) + np.pi/30):
    w = []
    
    for theta in (np.linspace(0, 2*np.pi, 500)):
        
        x = A * np.cos(theta) * np.cos(alph)
        y = A * np.cos(theta) * np.sin(alph)
        z = C * np.sin(theta)
        
        w.append([x, y, z])
        
    key = 'alph_{}'.format(ctr)
    
    wires[key] = [np.array(w), {'c': 'g'}]
```

```python
# remove wire sections where z < z_0
for k, (w, kwargs) in wires.items():
    w[w[:, 2] < Z_0] = np.nan
    wires[k] = [w, kwargs]
```

```python
P_E = np.array([5.4, 1, -.5])
V_E = np.array([-1, -.1, .2])

phi = 25
a = 140
b = 130

ax = plot_wires_3d(wires, p_e=P_E, v_e=V_E)[1]
ax.set_xlim(-7, 7)
ax.set_ylim(-7, 7)
ax.set_zlim(0, 14)

wires_transformed = to_mpp(
    wires, p_e=P_E, v_e=V_E, phi=phi, a=a, b=b, width=15)[2]
```
